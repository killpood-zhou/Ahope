```python
import re
import requests
from pprint import pprint

url="https://kyfw.12306.cn/otn/resources/js/framework/station_name.js?station_version=1.9044"
text=requests.get(url,verify=False)
stations=re.findall(r'([A-Z]+)\|([a-z]+)',text.text)
ac=dict(stations)
add=dict(zip(ac.values(),ac.keys()))
#print(text.text)
pprint(add,indent=4)
```
